```python
participant_grouped = marriage_postal.groupby(['electoral_division']).sum()['participant']
```

```python
eligible_grouped = marriage_postal.groupby(['electoral_division']).sum()['eligible']
```

```python
# creating a new dataframe
grouped_df = pd.DataFrame({'sum_participant': participant_grouped, 'sum_eligible': eligible_grouped})
```

```python
grouped_df
```

```python
df = pd.read_csv('electorate-results.csv')
```

```python
electorate_dataframe = df
```

```python
# merging two dataframe by right merge
merged_dataframes = pd.merge(grouped_df, electorate_dataframe, how='right', on=['electoral_division', 'electoral_division'])
```
