```python
# Read in electorate division info csv
div_df = pd.read_csv("01-output_electoral_division/electoral_division.csv")
div_df.head()
```

```python
# Add in division_id column to cultural_diversity df through left join
merged = pd.merge(div_df, age_df, how="inner", on="electoral_division")
merged.head()
```

```python
merged.head()
```

```python
# Read in Excel and specific tab population count
pop_df = pd.read_excel("Resources/commonwealth electorate data.xls", sheet_name = "Table 1")
```

```python
# Check datatypes
pop_df.dtypes
```

```python
# Count rows pop count
len(pop_df)
```

```python
# Summary statistics pop_df
pop_df.describe()
```

```python
# Merge dataframes to add in pop count
merged2 = pd.merge(merged, pop_df,  
                     on ='electoral_division',  
                     how ='inner') 

merged2
```

```python
# Delete electoral_division and state column
merged2.drop(["state", "electoral_division"], axis=1, inplace=True)
merged2.head()
```

```python
# Rename columns
merged2.rename(columns={"0 to 17 years old": "0-17_percent", "18 to 34 years old": "18-34_percent", "35 to 49 years old":"35-49_percent", "50 to 64 years old":"50-64_percent", "65 to 79 years old":"65-79_percent", "80 years old and over":"80+_percent", "2017 Estimated Resident Population counts":"pop_count_est"}, inplace=True)
merged2.head()
```

```python
# Change percentages to whole numbers
merged2["0-17_percent"] = merged2["0-17_percent"] * 100
merged2["18-34_percent"] = merged2["18-34_percent"] * 100
merged2["35-49_percent"] = merged2["35-49_percent"] * 100
merged2["50-64_percent"] = merged2["50-64_percent"] * 100
merged2["65-79_percent"] = merged2["65-79_percent"] * 100
merged2["80+_percent"] = merged2["80+_percent"] * 100
```

```python
# Set a comma separator
merged2["pop_count_est"].apply(lambda x : "{:,}".format(x))
```

```python
merged2
```

```python
df_age_div_pop = merged2.set_index("division_id")
```
